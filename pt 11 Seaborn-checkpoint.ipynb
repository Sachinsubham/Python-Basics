# Seaborn

It is used for making stastical graph in python. It builds on top of matplotlib and integrates closely with pandas data structure.

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
var1 = [1,2,3,4,6,8]
var2 = [8,9,7,3,2,4]

# By Matplotlib

plt.plot(var1,var2)
plt.show()
```

```python
# By seaborn

sns.lineplot(x =var1,y=var2)
plt.show()
```

```python
var1 = [1,2,3,4,6,8]
var2 = [8,9,7,3,2,4]

import pandas as pd

df = pd.DataFrame({"var1":var1,"var2":var2}) 
sns.lineplot(x ="var1",y="var2",data = df)
plt.show()
```

```python
# taking real example to unserstand seaborn

df1 = sns.load_dataset("penguins")                # import data from the google nd github
df1
```

```python
sns.lineplot(x ="bill_length_mm",y ="flipper_length_mm",data = df1)
plt.show()
```

```python
sns.lineplot(x ="bill_length_mm",y ="flipper_length_mm",data = df1 , hue ="sex")
plt.show()
```

```python
sns.lineplot(x ="bill_length_mm",y ="flipper_length_mm",data = df1 , hue ="sex",palette ="rocket_r")   # "Palette" is for colors
plt.show()
```

```python
sns.lineplot(x ="bill_length_mm",y ="flipper_length_mm",data = df1 , hue ="sex", style="sex", palette ="rocket_r", markers =["o",">"])
plt.show()
```

# Bar graph

```python
df1 = sns.load_dataset("penguins")
df1
```

```python
sns.barplot(x = "island",y = "bill_length_mm",data =df1)
plt.show()
```

```python
sns.barplot(x = df1.island, y = df1.bill_length_mm)
plt.show()
```

```python
sns.barplot(x = "island",y = "bill_length_mm",data =df1, hue = "sex")
plt.show()
```

```python
order1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x = "island",y = "bill_length_mm",data =df1, hue ="sex",order = order1)
plt.show()
```

```python
order1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x = "island",y = "bill_length_mm",data =df1, hue ="sex",order = order1, hue_order = ["Female","Male"])
plt.show()
```

# Histogram

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1 = sns.load_dataset("penguins")
df1
```

```python
sns.displot(df1["flipper_length_mm"])
plt.show()
```

```python
sns.histplot(df1["flipper_length_mm"])
plt.show()
```

```python
sns.histplot(df1["flipper_length_mm"], bins = [170,180,190,200,210,220,230])
plt.show()
```

```python
sns.histplot(df1["flipper_length_mm"], bins = [170,180,190,200,210,220,230], color = "red")
plt.show()
```

Kde (Kernel Density Estimate) - 

```python
sns.histplot(df1["flipper_length_mm"], bins = [170,180,190,200,210,220,230], kde = True)   
plt.show()  
```

```python
sns.histplot(df1["flipper_length_mm"], bins = [170,180,190,200,210,220,230], kde = False)
plt.show()
```
