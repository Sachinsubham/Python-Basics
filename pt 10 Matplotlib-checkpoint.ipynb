# Scatter Plot

```python
x =[1,4,6,3,8]
y =[5,2,8,7,1]
plt.title("scatter plot")      # for graph title
plt.xlabel("number")           # to give axis name
plt.ylabel("month")
plt.scatter(x,y)
plt.show()
```

```python
x =[1,4,6,3,8]
y =[5,2,8,7,1]
plt.title("scatter plot")     
plt.xlabel("number")       
plt.ylabel("month")
c = ["red","green","brown","yellow","blue"]
plt.scatter(x,y,color = c, s = 150)
plt.show()
```

# Image modification

PIL (python imaging library) - It is an open-source library for image processing tasks that requires python programming language. PIL can perform tasks on an image such as reading, rescaling, saving in different image formats.

```python
# importing libraries

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image     # importing PIL

fname = r'bird.jpg' 

# opening of image using PIL

image = Image.open(fname).convert('L')

# maping image to gray scale

plt.imshow(image ,cmap = 'rainbow') 
plt.show()
```

```python
# importing libraries

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image     # importing PIL

fname = r'salman khan.jpg' 

# opening of image using PIL

image = Image.open(fname).convert('L')

# maping image to gray scale

plt.imshow(image ,cmap = 'gray')
plt.title ('Salman Khan')
plt.xlabel("length")
plt.ylabel("breadth")
plt.show()
```
